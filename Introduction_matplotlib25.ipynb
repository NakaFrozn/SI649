# Introduction to Matplotlib


## Overview

matplotlib is a python library that is useful for making static visualizations, including scientific visualizations.

Rather than using a Grammar of Graphics approach (which we will utilize with Altair), matplotlib can be used to make specific types of visualizations (scatterplots, barplots, etc.).

It offers many options and can be highly customized, although it can sometimes be difficult to figure out how to do something specific. The documentation and sites like stackoverflow are helpful here.

Additional tutorials are available here: https://matplotlib.org/stable/tutorials/index, https://jakevdp.github.io/PythonDataScienceHandbook/04.00-introduction-to-matplotlib.html

## A simple example

Let's first create a simple dataset and make a figure! Here are the basic components of a matplotlib visualization:

**Figure:** The entire plot or canvas for the visualization.

**Axes:** The area/pane within a figure where you plot the data. You can have multiple axes in one figure.

**Line & Marks:** Data representations

**Axis:** X & Y, with limits and ticks representing data

**Labels:** Labels for X & Y axes, labels for data points

**Title:** Text at the top of the plot

````python
# Create some data
x = [4.0, 5.0, 6.0, 7.0, 8.0, 9.0, 10.0, 11.0, 12.0, 13.0, 14.0]
y = [4.26, 6.95, 7.24, 4.82, 9.96, 7.58, 8.81, 8.04, 8.33, 10.84, 5.68,]

# Create a plot
fig, ax = plt.subplots()

print('Figure type: ',type(fig))
print('Ax type: ',type(ax))

# Plot the data on the axis as a scatter plot
ax.scatter(x, y)

# Show the plot -- What happens if you comment this out?
plt.show()
````

## Creating plots with multiple axes

````python
# Create a plot with multiple panes (2 rows)
fig, axes = plt.subplots(nrows=2)

print('Axes type: ',type(axes))
print('Individual ax type: ',type(axes[0]))

plt.show()
````

````python
fig, axes = plt.subplots(nrows=2)

# plot the data on the first axis
ax = axes[0]
ax.scatter(x, y)

# plot it as a bar plot on the second axis
ax = axes[1]
ax.bar(x, y)

plt.show()
````

````python
# Link the panes together (use the same x-axis)
fig, axes = plt.subplots(nrows=2, sharex=True)

# plot the data on the first axis
ax = axes[0]
ax.scatter(x, y)

# plot it as a bar plot on the second axis
ax = axes[1]
ax.bar([val * 2 for val in x], y)

plt.show()
````

## What's missing?


**Plot aesthetics!** (figure size, symbols, labels, tickmarks, and legends)

````python
# Create a square figure
fig, ax = plt.subplots(figsize=(5, 5))

# Change the size and color of the scatter plot
ax.scatter(x, y, s=100, color='magenta', marker='*')

# For more colors, see: https://matplotlib.org/stable/gallery/color/named_colors.html
# you can also use hex codes like color="#058ED9" or RGB values color=[0.8, 0.5, 0.1]

# For marker styles, see: https://matplotlib.org/stable/api/markers_api.html

plt.show()
````

````python
fig, ax = plt.subplots()

ax.scatter(x, y)

# Change the limits of the y-axis
ax.set_ylim(0, 13)

# Change the tick-marks on x-axis
ax.set_xticks([4, 9, 14])

# Add a label on the x-axis
ax.set_xlabel("x-axis")

# Add a title:
ax.set_title("Title")


# Show the plot
plt.show()
````

## Adding text labels to plot

````python
fig, ax = plt.subplots()

ax.scatter(x, y)

# Add a big text label for the second point (note that it is not jittered)
ax.text(x[1]+0.5, y[1]+0.5, 'text label', size=20)

plt.show()
````

## Additional dimensions

Color & size can encode additional information.

````python
# Add a third dimension of data
z = np.array([10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0])

fig, axes = plt.subplots(nrows=2, figsize=(4, 8))

# make a scatter plot, use the z variable to control size
ax = axes[0]
ax.scatter(x, y, s=z*8)

# or use the z variable to control color
ax = axes[1]
ax.scatter(x, y, c=z)

plt.show()
````

# Line plots, legends, and default colors

````python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 5, 2, 5, 2, 5, 2]
z = [2, 4, 5, 4, 1, 3, 6, 4]

fig, ax = plt.subplots()

# The default plot is a line plot
ax.plot(x, y)
ax.scatter(x, y)

plt.show()
````

````python
fig, ax = plt.subplots()

# We can name the lines using "label"
ax.plot(x, y, label='y')

# We can also put additional series on the same axis (and use a different line-style)
ax.plot(x, z, label='z', linestyle=':')

# It will automatically change the color, which we can access as C0, C1, C2, ...

# Add a text detail in the second color
ax.text(x[-2], z[-2], 'detail', color='C1')

# The line names will automatically show up in the legend, which we place in a corner
ax.legend(loc='upper left')

plt.show()
````

## Working with data

For this lab, we will be looking at the [Taylor Swift Spotify Dataset](https://www.kaggle.com/datasets/jarredpriester/taylor-swift-spotify-dataset?resource=download) from Kaggle.

````python
# Load the Taylor Swift data
df = pd.read_csv('taylor_swift_spotify.csv', index_col=0)
df.head()
````

````python
# plot energy by acousticness
x = df['acousticness'].values
y = df['energy'].values

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x, y)

ax.set_xlabel('Acousticness')
ax.set_ylabel('Energy')

# add a dashed line for the diagonal
ax.plot([min(x), max(x)], [max(x), min(x)], linestyle='--', color='k', alpha=0.5)

plt.show()
````

````python
# plot energy by acousticness
x = df['acousticness'].values
y = df['energy'].values

fig, ax = plt.subplots(figsize=(6,6))
ax.scatter(x, y)

ax.set_xlabel('Acousticness')
ax.set_ylabel('Energy')

# add labels for outliers
titles = df['name'].values
for i, yi in enumerate(y):
    if yi > .66 and x[i] > .66:
        ax.text(x[i], y[i], titles[i])

plt.show()
````

````python
# Get the albums, dropping nans
albums = df['album'].dropna().values

album_counter = Counter(albums)
n_albums = len(album_counter)

# get a sorted list of albums, in reverse alphabetical order
list_of_albums = sorted(album_counter)[::-1]

# get the count for each album
album_counts = [album_counter[g] for g in list_of_albums]

# Make a bar plot of albums counts
fig, ax = plt.subplots(figsize=(5, 6))

ax.barh(range(n_albums), album_counts)

ax.set_yticks(range(n_albums))
ax.set_yticklabels(list_of_albums)

# add text labels for the counts  (with some minor tweaks)
for i, g in enumerate(list_of_albums):
    ax.text(album_counter[g]+2, i-.3, str(album_counter[g]), size=10)

# fix the plot limits
ax.set_xlim(0, 70)

ax.set_title('Number of songs per album')

plt.show()
````

## Bonus Content: XKCD Plot

Matplotlib has some built in styles, like [XKCD](https://xkcd.com/)'s signature style. For more information: [read the documentation](https://matplotlib.org/stable/gallery/showcase/xkcd.html#sphx-glr-gallery-showcase-xkcd-py).

````python
# disable warning messages for fonts not found
import logging
logging.getLogger('matplotlib.font_manager').disabled = True

# get the year from the data
df['year']=pd.to_datetime(df['release_date'],format='%Y-%m-%d').dt.year
# group by the year
songs_by_date = df.groupby('year')['name'].count().reset_index()

# plot in XKCD style
with plt.xkcd():
    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(songs_by_date['year'],songs_by_date['name'])
    ax.set_ylabel('Number of Songs')
    ax.set_xlabel('Year')
    ax.set_title('Taylor Swift Songs Released per Year')

    # set x ticks specifically & rotate
    ax.set_xticks([x for x in range(min(songs_by_date['year'].values),max(songs_by_date['year'].values))])
    ax.set_xticklabels(ax.get_xticklabels(), rotation=75)

    # add an annotation with an arrow
    ax.annotate(
        'Release of Evermore\nand Folklore albums',
        xy=(2020, 82), arrowprops=dict(arrowstyle='->'), xytext=(2011, 80))

plt.show()
````
